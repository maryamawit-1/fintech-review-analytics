```python
import pandas as pd

clean_df = pd.read_csv(
    "../data/processed/bank_reviews_clean.csv"
)

sentiment_df = pd.read_csv(
    "../data/processed/task2_sentiment_analysis.csv"
)

sentiment_df['date'] = clean_df['date']
sentiment_df['source'] = clean_df['source']

sentiment_df = sentiment_df[
    [
        'review',
        'rating',
        'date',
        'bank',
        'source',
        'sentiment_label',
        'sentiment_score',
        'identified_theme'
    ]
]

sentiment_df.to_csv(
    "../data/processed/task2_sentiment_analysis.csv",
    index=False
)

print("Updated CSV saved successfully!")
```
